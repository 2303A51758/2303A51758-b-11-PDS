```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (make sure daily_temperature.csv has columns: Day, Temperature)
data = pd.read_csv("daily_temperature.csv")

# Plotting
plt.figure(figsize=(10,6))
plt.plot(data['Day'], data['Temperature'], marker='o', linestyle='-', color='b')

# Add labels, title, and grid
plt.xlabel("Day of the Month")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature Variation Over a Month")
plt.grid(True)

# Show plot
plt.show()

# Find hottest and coldest day
hottest_day = data.loc[data['Temperature'].idxmax()]
coldest_day = data.loc[data['Temperature'].idxmin()]

print(f"🔥 Hottest Day: Day {hottest_day['Day']} with {hottest_day['Temperature']}°C")
print(f"❄️ Coldest Day: Day {coldest_day['Day']} with {coldest_day['Temperature']}°C")
```

```python
# Create a dummy daily_temperature.csv file for demonstration
data = {'Day': range(1, 31),
        'Temperature': [20 + i*0.5 + (i%5)*2 - (i%7)*1 for i in range(30)]}
dummy_df = pd.DataFrame(data)
dummy_df.to_csv("daily_temperature.csv", index=False)

print("Dummy 'daily_temperature.csv' created.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (sales_data.csv should have columns: Quarter, Sales)
data = pd.read_csv("sales_data.csv")

# Plotting
plt.figure(figsize=(8,5))
plt.plot(data['Quarter'], data['Sales'], marker='o', linestyle='-', color='g', label="Sales")

# Highlight peak sales with red marker
peak = data.loc[data['Sales'].idxmax()]
plt.scatter(peak['Quarter'], peak['Sales'], color='red', s=100, label="Peak Sales")

# Add labels, title, legend, and grid
plt.xlabel("Quarter")
plt.ylabel("Sales (in USD)")
plt.title("Quarterly Sales Performance")
plt.legend()
plt.grid(True)

# Show plot
plt.show()

# Print peak sales info
print(f"📈 Peak Sales: {peak['Sales']} in {peak['Quarter']}")
```

```python
# Create a dummy sales_data.csv file for demonstration
data = {'Quarter': ['Q1', 'Q2', 'Q3', 'Q4'],
        'Sales': [15000, 18000, 22000, 20000]}
dummy_df = pd.DataFrame(data)
dummy_df.to_csv("sales_data.csv", index=False)

print("Dummy 'sales_data.csv' created.")
```
