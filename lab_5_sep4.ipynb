```python
import pandas as pd

# -----------------------------
# Step 1: Create Sample Datasets
# -----------------------------
# Products dataset
products = pd.DataFrame({
    'product_id': [101, 102, 103, 104, 105],
    'product_name': ['Laptop', 'Smartphone', 'Tablet', 'Headphones', 'Smartwatch'],
    'category': ['Electronics', 'Electronics', 'Electronics', 'Accessories', 'Electronics'],
    'price': [70000, 30000, 20000, 5000, 15000]
})

print("Products Dataset:")
print(products, "\n")

# Sales dataset
sales = pd.DataFrame({
    'sale_id': [1, 2, 3, 4, 5, 6, 7],
    'product_id': [101, 102, 103, 104, 101, 105, 106],  # 106 doesnâ€™t exist in products
    'quantity': [3, 5, 2, 10, 1, 4, 2]
})

print("Sales Dataset:")
print(sales, "\n")

# -----------------------------
# Step 2: Merge Datasets
# -----------------------------
# Inner Join
merged_inner = pd.merge(sales, products, on="product_id", how="inner")
print("Inner Join Result:")
print(merged_inner, "\n")

# Outer Join
merged_outer = pd.merge(sales, products, on="product_id", how="outer")
print("Outer Join Result:")
print(merged_outer, "\n")

# -----------------------------
# Step 3: Compute Statistics
# -----------------------------
# Add revenue column
merged_inner['revenue'] = merged_inner['quantity'] * merged_inner['price']
print("Merged Data with Revenue:")
print(merged_inner, "\n")

# Overall statistics
total_revenue = merged_inner['revenue'].sum()
avg_price = products['price'].mean()
product_sales = merged_inner.groupby('product_name')['quantity'].sum()

print("Total Revenue:", total_revenue)
print("Average Product Price:", avg_price, "\n")
print("Product-wise Total Sales:")
print(product_sales, "\n")

# Category-wise statistics
category_sales = merged_inner.groupby('category')['revenue'].sum()
best_category = category_sales.idxmax()

print("Category-wise Revenue:")
print(category_sales, "\n")
print("Best-Selling Category (by revenue):", best_category)
```

```python
import pandas as pd

# -----------------------------
# Step 1: Create Sample Datasets
# -----------------------------
# Teachers dataset
teachers = pd.DataFrame({
    'teacher_id': [1, 2, 3, 4],
    'teacher_name': ['Alice', 'Bob', 'Charlie', 'David'],
    'department': ['CS', 'Math', 'CS', 'Physics']
})

print("Teachers Dataset:")
print(teachers, "\n")

# Courses dataset
courses = pd.DataFrame({
    'course_id': [101, 102, 103, 104, 105, 106],
    'course_name': ['AI', 'Data Structures', 'Calculus', 'Physics I', 'ML', 'Algorithms'],
    'teacher_id': [1, 1, 2, 4, 1, 3]  # some teachers handle multiple courses
})

print("Courses Dataset:")
print(courses, "\n")

# -----------------------------
# Step 2: Merge Datasets
# -----------------------------
# Inner Join (only matching teacher_ids)
merged_inner = pd.merge(courses, teachers, on="teacher_id", how="inner")
print("Inner Join Result:")
print(merged_inner, "\n")

# Left Join (all courses, even if teacher missing)
merged_left = pd.merge(courses, teachers, on="teacher_id", how="left")
print("Left Join Result:")
print(merged_left, "\n")

# -----------------------------
# Step 3: Compute Statistics
# -----------------------------
# Number of courses per teacher
courses_per_teacher = merged_inner.groupby('teacher_name')['course_id'].count()
avg_courses_per_teacher = courses_per_teacher.mean()

print("Courses per Teacher:")
print(courses_per_teacher, "\n")
print("Average Courses per Teacher:", avg_courses_per_teacher, "\n")

# Workload by department (total number of courses taught by dept)
department_workload = merged_inner.groupby('department')['course_id'].count()

print("Department-wise Workload:")
print(department_workload, "\n")

# Best performing department (highest workload)
busiest_department = department_workload.idxmax()
print("Department with Highest Workload:", busiest_department)
```
